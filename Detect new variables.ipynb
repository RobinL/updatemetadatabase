```python
#Connect to the metadatabase using pyodbc
import os
import pyodbc
import re
dst = os.path.join(os.getcwd(),r"metadatabase.mdb")
MDB = dst
DRV = '{Microsoft Access Driver (*.mdb, *.accdb)}'

con = pyodbc.connect('DRIVER=%s;DBQ=%s;' % (DRV,MDB))
cur = con.cursor()
```

```python
import pandas as pd
import pandas.io.sql as sql 
pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', 500)
```

```python
#Import the list variables in the new datasets
src =r"variablenamesandformats/List of variables in forms.xls"
olddf = pd.read_excel(src,"old _forms",header=1)
newdf = pd.read_excel(src,"new_forms",header=1)
```

```python
#filter from olddf any variables which are in newdf.  We want descriptions from the new database to override descriptions from the old db

not_in = -olddf['Variable'].isin(newdf["Variable"])
olddf = olddf[not_in]


```

```python
#concatenate the two datasets
alldf = pd.concat([olddf, newdf])
distinctnewvars = alldf.groupby(["Variable","Label"]).size().reset_index()
distinctnewvars =distinctnewvars.rename(columns= {0: "shouldbe1"})
```

```python
#Create a dataframe of the variables in the meta database
variables = sql.read_frame('select * from tblVariables', con)
```

```python
#Find variables which are in distinct new variables but are not in the metadatabase
not_in = -distinctnewvars['Variable'].isin(variables["OriginalORSName"])
distinctnewvars = distinctnewvars[not_in]
```

```python
#Now we have a list of the distinct new variables

#Figure out there are groups of variables in the list of new variables
distinctnewvars["LabelNoCode"] = distinctnewvars["Label"].str.match("([^#]+)").str.get(0).str.strip()
```

```python
groups = distinctnewvars.groupby("LabelNoCode").size().reset_index()
groups =groups.rename(columns= {0: "sizeofgroup"})
groups
```

```python
#Tomorrow join these back on to the new variables table and attempt to detect the prefix and suffix

#The difficulty is going to be detecting whether each variable needs a new description grou
#I guess you just look in the table to see whether the prefix plus basename exists.  If it does, you don't need a new description group
#you just put the variable in and use its current descirption group

distinctnewvarswithgroups = distinctnewvars.merge(groups, left_on = "LabelNoCode", right_on = "LabelNoCode", how="left")
```

```python
#need to create
#suffix
#suffixcode
#prefix
#basename
#revisionnum
import re
def revisionnum(x):
    r = re.search("_\d$",x)
    if r:
        return r.group(0)
    else:
        return ""

distinctnewvarswithgroups["RevisionNum"] = distinctnewvarswithgroups["Variable"].apply(revisionnum)

```

```python
#Prefix

#Get list of prefixes
prefixes = sql.read_frame('select * from tblPrefix', con)
plist =  prefixes["ORSPrefix"].tolist()
plist = list(filter(None, plist))
reexp = "|".join(plist)
reexp = "(" + reexp + ")(.+)"

def prefix(x):
    
    r = re.match(reexp,x)
    if r:
        prefix = r.group(1)
    else:
        prefix = ""
        
    return prefix

distinctnewvarswithgroups["prefix"] = distinctnewvarswithgroups["Variable"].apply(prefix)
```

```python
from functools import partial
def basenamecandidate(row):
   
    x = row["Variable"]
    #Remove revisionnum if it exists
    r = re.search("(.+)_\d$",x)
    if r:
        x = r.group(1)
    
    #Remove prefix if it exists
    r = re.match(reexp,x)
    if r:
        x = r.group(2)
  
    
    return x

distinctnewvarswithgroups["basenamecandidate"] = distinctnewvarswithgroups.apply(basenamecandidate, axis=1)

def grouplen(minus,row):
    #Find out what the base name 'would be' if there was a suffix, and check whether we have a group
    #Get the max length of the current basename+suffix string.  Take two off it.  Then truncate this string if need be
    thisgroup = distinctnewvarswithgroups[distinctnewvarswithgroups["LabelNoCode"]==row["LabelNoCode"]]
    maxlen = thisgroup["basenamecandidate"].str.len().max()
    maxlenofbase = maxlen-minus
    x = row["basenamecandidate"][:maxlenofbase]
    x = row["prefix"] + x + row["RevisionNum"]
    return x



distinctnewvarswithgroups["grouplen1txt"] = distinctnewvarswithgroups.apply(partial(grouplen,1), axis=1)
distinctnewvarswithgroups["grouplen2txt"] = distinctnewvarswithgroups.apply(partial(grouplen,2), axis=1)

def isitagroup(field,row):
    thisgroup = distinctnewvarswithgroups[distinctnewvarswithgroups["LabelNoCode"]==row["LabelNoCode"]]
    a = thisgroup.groupby(field).size().reset_index()
    if len(a.index) ==1:
        if a[0].min()>1:
            return True
        else:
            return False
    else:
        return False

    

distinctnewvarswithgroups["grouplen1"] = distinctnewvarswithgroups.apply(partial(isitagroup,"grouplen1txt"), axis=1)
distinctnewvarswithgroups["grouplen2"] = distinctnewvarswithgroups.apply(partial(isitagroup,"grouplen2txt"), axis=1)
```

```python
def getsuffix(row):
   
    x = row["Variable"]
    #Remove revisionnum if it exists
    r = re.search("(.+)_\d$",x)
    if r:
        x = r.group(1)
    
    #Remove prefix if it exists
    r = re.match(reexp,x)
    if r:
        x = r.group(2)
        
    if row["grouplen1"]:
        return x[-1:]
    elif row["grouplen2"]:
        thisgroup = distinctnewvarswithgroups[distinctnewvarswithgroups["LabelNoCode"]==row["LabelNoCode"]]
        maxlen = thisgroup["Variable"].str.len().max()
        if len(row["Variable"]) < maxlen:
            return x[-1:]
        else:
            return x[-2:]
    else:
        return ""
  
    return x

distinctnewvarswithgroups["suffix"] = distinctnewvarswithgroups.apply(getsuffix, axis=1)
```

```python
#Now we just need to get the basename, - anything that isn't the prefix, the suffix or the revisionnum
def getbasename(row):
    lp =len(row["prefix"])
    ls =len(row["suffix"]) + len(row["RevisionNum"])
    if ls == 0:
         return row["Variable"][lp:]
    else:
        return row["Variable"][lp:-ls]

distinctnewvarswithgroups["basename"] = distinctnewvarswithgroups.apply(getbasename, axis=1)

#Check that it's worked
distinctnewvarswithgroups["vcheck"] = distinctnewvarswithgroups["prefix"]\
+ distinctnewvarswithgroups["basename"]\
+ distinctnewvarswithgroups["suffix"]\
+ distinctnewvarswithgroups["RevisionNum"]

#print(distinctnewvarswithgroups["vcheck"] == distinctnewvarswithgroups["Variable"])
distinctnewvarswithgroups[["Variable","basename","prefix","RevisionNum","suffix","LabelNoCode"]]
```

```python
#Now these need to be added to the database!!
# SQLUpdateVariablesTable = """
# Insert into tblVariables (originalorsname, basename,)
# values (?,?,?)
# """


SQLFindPrefix = """
select id, finalprefix from 
tblPrefix
where ORSPrefix = ?
"""

SQLFindDescCode = """
select groupdescriptionid from 
tblVariables
where
basename = ? and prefixid = ? and revisionnum = ?
"""

SQLInsertDescription = """
insert into tblDescriptiongroups 
(description)
values (?)
"""

SQLInsertVariable = """
insert into tblVariables 
(originalORSname, basename, finalname, prefixid, suffix, suffixcode, revisionnum, groupdescriptionid, inORSRawDataset)
values (?,?,?,?,?,?,?,?,?)
"""

SQLGetSuffix = """
select suffixcode
from tblVariables
where suffix = ?
"""

listofNewVarIds=[]

for row in distinctnewvarswithgroups.iterrows():
    row = row[1]
    basename = row["basename"]
    prefix = row["prefix"]
    RevisionNum = row["RevisionNum"]
    variable = row["Variable"]
    suffix = row["suffix"]
    description = row["LabelNoCode"]
    
    if suffix:
        suffixget = cur.execute(SQLGetSuffix,(suffix))
        suffixcode = str(suffixget.fetchone()[0])
    else:
        suffixcode = None
    
    #Figure out the relevant prefix ID and get the finalprefix
    records = cur.execute(SQLFindPrefix, (prefix))
    idrecord = records.fetchone()
    if idrecord:
        prefixid = str(idrecord[0])
        finalprefix = str(idrecord[1])
    else:
        prefixid = "14"
        finalprefix = None
     
    #If this basename, prefix and revision num combo already in dataset, figure out the description group
    records = cur.execute(SQLFindDescCode, (basename, prefixid, RevisionNum))
    
  
    iddesc = records.fetchone()
    
    if iddesc:
        
        descid = iddesc[0]
    else:
        descid = None
   

    #Otherwise add this description as a description group
    if descid == None:
        cur.execute(SQLInsertDescription,(description))
        cur.commit()
        descid = cur.execute("SELECT @@Identity").fetchall()[0][0]
        
    
    
    if prefixid:
        prefixid = str(prefixid)
    else:
        prefixid = "14"
    
    if finalprefix == None:
        finalprefixstr = ""
    else:
        finalprefixstr = finalprefix
    finalname = finalprefixstr + basename + suffix + RevisionNum
    descid = int(descid)
    
   
    #Write the new finalname and add to db   
    
    
    cur.execute(SQLInsertVariable, (variable, basename, finalname,str(prefixid), suffix,suffixcode, RevisionNum,str(descid), True))
    cur.commit()
    listofNewVarIds.append(cur.execute("SELECT @@Identity").fetchall()[0][0])
 
  
    
    
```

```python
#Now need to update the table tblVariablesForms so that the new variables are associated with the correct form ids

#First update the tblForms - add a field called prefix ID so we know which prefix the form is associated with

##Join on id of form
variables = sql.read_frame('select * from tblVariables', con)
newvars = variables[variables["ID"].isin(listofNewVarIds)]
```
