```python
import pandas as pd
import pandas.io.sql as sql 
pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
```

```python
#Import the list variables in the new datasets
src =r"variablenamesandformats/List of variables in forms.xls"
olddf = pd.read_excel(src,"old _forms",header=1)
newdf = pd.read_excel(src,"new_forms",header=1)
```

```python
#filter from olddf any variables which are in newdf

not_in = -olddf['Variable'].isin(newdf["Variable"])
olddf = olddf[not_in]
olddf

```

```python
#concatenate the two datasetsei
alldf = pd.concat([olddf, newdf])
distinctnewvars = alldf.groupby(["Variable","Label"]).size().reset_index()
distinctnewvars.head()
```
